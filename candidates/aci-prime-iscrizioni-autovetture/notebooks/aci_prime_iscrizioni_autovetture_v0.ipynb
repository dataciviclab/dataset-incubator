```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
```

```python
import os
os.chdir('../../..')

# Carica dati clean
df = pd.read_parquet('out/data/clean/aci_prime_iscrizioni_autovetture/2024/aci_prime_iscrizioni_autovetture_2024_clean.parquet')print(f'Righe totali: {len(df)}')
```

```python
print('Colonne:', list(df.columns))
print()
print(df.dtypes)
print()
print('Anni:', sorted(df['anno'].unique()))
print('Alimentazioni:', sorted(df['alimentazione'].unique()))
print('Regioni/province:', df['provincia'].nunique())
```

```python
# Totali per anno
annuale = df.groupby('anno')['prime_iscrizioni'].sum()
print(annuale.to_string())

plt.figure(figsize=(12, 4))
sns.barplot(x=annuale.index, y=annuale.values)
plt.title('Immatricolazioni auto nuove per anno (2017-2024)')
plt.ylabel('Prime iscrizioni')
plt.show()
```

```python
# Per tipo alimentazione
alim = df.groupby(['anno', 'alimentazione'])['prime_iscrizioni'].sum().reset_index()

plt.figure(figsize=(14, 5))
sns.lineplot(data=alim, x='anno', y='prime_iscrizioni', hue='alimentazione', marker='o')
plt.title('Immatricolazioni per tipo alimentazione (2017-2024)')
plt.ylabel('Prime iscrizioni')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Quota elettrico + ibrido sul totale
elettrificato = alim[alim['alimentazione'].str.contains('Elettrica|Ibrido BE|Ibrido GE', na=False)]
elettr_annuale = elettrificato.groupby('anno')['prime_iscrizioni'].sum()
quota = (elettr_annuale / annuale * 100).reset_index()
quota.columns = ['anno', 'quota_perc']
print(quota.to_string(index=False))

plt.figure(figsize=(10, 4))
sns.lineplot(data=quota, x='anno', y='quota_perc', marker='o')
plt.title('Quota elettrico+ibrido sulle immatricolazioni (%)')
plt.ylabel('%')
plt.show()
```
